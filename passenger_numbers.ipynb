```python
import seaborn as sns
import matplotlib.pyplot as plt

flights = sns.load_dataset('flights')
flights_pivot = flights.pivot(index='month', columns='year', values='passengers')
plt.rcParams["font.family"] = "monospace"
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(flights_pivot,
    cmap='RdPu',
    annot=False,
    cbar=True,
    linewidth=2,
    center=flights_pivot.loc['Jan', 1955]
    )

plt.title('Passenger number variations', pad=20, weight='bold', fontsize=20)
plt.xlabel('Year', labelpad=20, fontsize=15)
plt.ylabel('Month', labelpad=20, fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.rcParams['figure.autolayout']
plt.show()
```
