```python
# inspected the excel file before & found that admission_type_ids 5, 6, and 8 are invalid. Must remove!

valid_admission_type_ids = [1, 2, 3, 4, 7]

df1 = df1[df1['admission_type_id'].isin(valid_admission_type_ids)]
df2 = df2[df2['admission_type_id'].isin(valid_admission_type_ids)]

# keeping only relevant columns in df1
df1 = df1[['time_in_hospital', 'admission_type_id', 'num_lab_procedures', 'age']]
```

## ANOVA

```python
# merging df1 and df2 on 'admission_type_id'

merged_df = pd.merge(df1, df2, on='admission_type_id', how='inner')
```

```python
import scipy.stats as stats

anova_df = df1[['time_in_hospital', 'age']]

groups = anova_df.groupby('age')['time_in_hospital'].apply(list)
f_statistic, p_value = stats.f_oneway(*groups.values)

print(f"F-statistic: {f_statistic:.2f}, \np-value: {p_value}")
```
